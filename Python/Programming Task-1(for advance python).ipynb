# 1470. Shuffle the Array

Given the array nums consisting of 2n elements in the form [x1,x2,...,xn,y1,y2,...,yn].

Return the array in the form [x1,y1,x2,y2,...,xn,yn]
 

Examle 1:

Input: nums = [2,5,1,3,4,7], n = 3
Output: [2,3,5,4,1,7] 
Explanation: Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3
,5,4,1,7].Example 2:

Input: nums = [1,2,3,4,4,3,2,1], n = 4
Output: [1,
4,2,3,3,2,41]
Example 3:

Input: nums = [1,1,2,2], n = 2
Output: [1,2,1,2]


```python
class Solution:
    def shuffle(self, nums, n):
       i = 0
       j = n
       list1 = []

       while j < len(nums):
        list1.append(nums[i])
        list1.append(nums[j])
        i+=1
        j+=1
        
       return list1

s1 = Solution()
result1,result2,result3 = s1.shuffle([2,5,1,3,4,7], 3) , s1.shuffle([1,2,3,4,4,3,2,1],4), s1.shuffle([1,1,2,2],2)

print(result1)
print(result2)
print(result3)
```
